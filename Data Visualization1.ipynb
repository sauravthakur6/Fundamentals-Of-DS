```python
data=pd.read_csv("Data/winemag-data-130k-v2.csv",index_col=0)
data.head(3)
```

```python
fig,ax=plt.subplots()
data1=data['points'].value_counts()
points=data1.index
frequency=data1.values
ax.bar(points,frequency)
ax.set_title("WRS")
ax.set_xlabel("Points")
ax.set_ylabel("Frequency")
```

```python
data.groupby("country").price.mean().sort_values(ascending=False)[:10].plot.bar()
```

```python
obj=("Python","C","Java","Pearl")
y_pos=np.arange(len(obj))
per=[10,9,6,6]
plt.bar(y_pos,per,width=0.8,align='edge',alpha=0.7,color=['red','blue','grey','orange'])
plt.xticks(y_pos,obj)
plt.ylabel("Usage")
plt.xlabel("Language")
plt.title("PLU")
```

```python
plt.figure(figsize=(12,7))
countries=['USA',"bRAZIL","RUSSIA",'SPAIN','UK']
totalDeaths=[112596,37312,5971,27136,40597]
plt.bar(countries,totalDeaths,width=0.9,align='center',color='cyan',edgecolor='red')
i=1.0
j=2000
for i in range(len(countries)):
    plt.annotate(totalDeaths[i],(-0.1+i,totalDeaths[i]+ j))
    plt.legend(labels=['Total Deaths'])
    plt.title("BAr plot representing")
    plt.xlabel("Countries")
    plt.ylabel("Deaths")
    plt.savefig("1BarPlot.png")
    plt.show()
```

```python
plt.figure(figsize=[15,10])
total_death=[15000,21000,3100,2000]
total_recovery=[77700,45000,60000,50000]
active_case=[70000,46000,4715,11100]
country=["USA","INDIA","UK","BRAZIL"]
X=np.arange(len(total_death))
plt.bar(X,total_death,color='r',width=0.25)
plt.bar(X+0.25,total_recovery,color='green',width=0.25)
plt.bar(X+0.5,active_case,color='orange',width=0.25)
plt.legend(["Total Deaths","Total Recoveries","Active Cases"])
plt.xticks([i+0.25 for i in range(5)],country)
plt.title("Covid-19")
plt.savefig("bar3.png")
plt.show()
```

```python
Data2={'Tasks':[300,500,700],
     'Task Type':["Tasks Pending",'Tasks Ongoing','Tasks Completed']
     }
df2=pd.DataFrame(Data2)
df2.set_index("Task Type",inplace=True)
df2.plot.pie(y='Tasks',figsize=(10,10),autopct='%1.1f%%',startangle=90);
```

```python
%matplotlib inline 
labels=["Civil","Mechanical","Chemical"]
sizes=[15,30,45]
explode=(0.1,0.1,0.4)
fig,ax=plt.subplots()
ax.pie(sizes,
      explode=explode,
      labels=labels,
      autopct="%1.1f%%",
      shadow=True,
      startangle=90)
ax.axis("equal")
ax.set_title("Engineering")
```
