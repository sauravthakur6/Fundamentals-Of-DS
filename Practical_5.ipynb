```python
def event_probability(n,s):
    return n/s
```

```python
pa=event_probability(2,6)
pb=event_probability(3,6)
p=pa*pb
print(p)
```

```python
pw=event_probability(5,10)
pb=event_probability(3,9)
pg=event_probability(2,8)
print(pw*pb*pg)
```

```python
import itertools
from itertools import product
```

```python
box_1=['g','b','y']
perm=[]
for p in itertools.product(box_1,repeat=3):
    perm.append(p)
perm
```

```python
box_1=['g','b']#permutation without repeatition
perm =itertools.permutations(box_1)
for i in list(perm):
    print(i)
```

```python
box_1=['g','b','y']
perm =itertools.permutations(box_1)
for i in list(perm):
    print(i)
```

```python
box_1=['g','b','y']
perm=[]
for p in itertools.product(box_1,repeat=2):
    perm.append(p)
perm
```

```python
box_2=['g','b','y']
perm =itertools.permutations(box_2,2)
for i in list(perm):
    print(i)
```

```python
from itertools import combinations_with_replacement
from itertools import combinations
```

```python
box_1=['g','b']
comb=combinations_with_replacement(box_1,2)
for i in list(comb):
    print(i)
```

```python
box_1=['g','b','y','w']
comb=combinations_with_replacement(box_1,2)
for i in list(comb):
    print(i)
```

```python
box_1=['g','b','y','w']
comb=combinations(box_1,2)#Without repeatition
for i in list(comb):
    print(i)
```
